# Model training from Scratch

## Introduction
In this notebook, we will demystify the "magic" behind machine learning model training. Instead of using high-level libraries like Scikit-Learn or XGBoost, we will build a **Linear Regression** model completely from scratch using only `numpy` for matrix operations.

We will cover:
1.  **Data Generation**: Creating a simple synthetic dataset.
2.  **Model Representation**: Defining the hypothesis function.
3.  **Cost Function**: Implementing Mean Squared Error (MSE).
4.  **Gradient Descent**: Calculating gradients and updating weights manually.
5.  **Training Loop**: Putting it all together to train the model.

### 1. Data Generation

Let's generate a simple linear dataset following the formula:
$$ y = 3x + 4 + \epsilon $$
Where \\(\epsilon\\) is some random Gaussian noise.

```python
import numpy as np
import matplotlib.pyplot as plt

# Set seed for reproducibility
np.random.seed(123)

# Generate 100 data points
X = 2 * np.random.rand(100, 1)
y = 3 * X + 4 + np.random.randn(100, 1)  # y = 3x + 4 + noise

# Visualize the data
plt.figure(figsize=(10, 6))
plt.scatter(X, y, color='blue', label='Data points')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Synthetic Linear Data')
plt.legend()
plt.show()
```

### 2. Model Initialization

A linear regression model makes a prediction \\(\hat{y}\\) using a weight \\(w\\) and a bias \\(b\\):
$$ \hat{y} = wx + b $$

We start by initializing these parameters randomly.

```python
# Initialize parameters randomly
w = np.random.randn(1)
b = np.random.randn(1)

print(f"Initial weight: {w[0]:.4f}")
print(f"Initial bias: {b[0]:.4f}")
```

### 3. Forward Pass & Cost Function

The **Cost Function** (or Loss Function) tells us how bad our model is performing. For regression, we typically use **Mean Squared Error (MSE)**:

$$ J(w, b) = \frac{1}{N} \sum_{i=1}^{N} (y_i - (wx_i + b))^2 $$

```python
def predict(X, w, b):
    return X * w + b

def compute_cost(X, y, w, b):
    N = len(y)
    predictions = predict(X, w, b)
    cost = (1/N) * np.sum((y - predictions) ** 2)
    return cost

initial_cost = compute_cost(X, y, w, b)
print(f"Initial Cost (MSE): {initial_cost:.4f}")
```

### 4. Gradient Descent

To improve the model, we need to update \\(w\\) and a bias \\(b\\) in the direction that reduces the cost. We do this by calculating the partial derivatives (gradients) of the cost function.

$$ \frac{\partial J}{\partial w} = -\frac{2}{N} \sum_{i=1}^{N} x_i (y_i - \hat{y}_i) $$
$$ \frac{\partial J}{\partial b} = -\frac{2}{N} \sum_{i=1}^{N} (y_i - \hat{y}_i) $$

Then we update the parameters:
$$ w_{new} = w_{old} - \alpha \cdot \frac{\partial J}{\partial w} $$
$$ b_{new} = b_{old} - \alpha \cdot \frac{\partial J}{\partial b} $$

Where $\alpha$ is the **learning rate**.

```python
def compute_gradients(X, y, w, b):
    N = len(y)
    predictions = predict(X, w, b)
    error = y - predictions
    
    # Derivatives
    dw = -(2/N) * np.sum(X * error)
    db = -(2/N) * np.sum(error)
    
    return dw, db
```

### 5. Training Loop

Now we run the gradient descent step iteratively for a number of epochs.

*Hint - this is a very simple example but good for experimentation. Feel free to adjust the learning rate and number of epochs and check your results, as they may be unexpected. Remember to reinitialize random weights before each run*

```python
learning_rate = 0.1
epochs = 100
cost_history = []

print("Starting training...")

for i in range(epochs):
    # 1. Calculate gradients
    dw, db = compute_gradients(X, y, w, b)
    
    # 2. Update parameters
    w = w - learning_rate * dw
    b = b - learning_rate * db
    
    # 3. Log cost
    cost = compute_cost(X, y, w, b)
    cost_history.append(cost)
    
    if i % 10 == 0:
        print(f"Epoch {i}: Cost {cost:.4f}, w={w[0]:.2f}, b={b[0]:.2f}")

print(f"\nFinal parameters: w={w[0]:.4f}, b={b[0]:.4f}")
print(f"Target parameters: w=3.0000, b=4.0000 (approx)")
```

### 6. Visualization

Let's see how our model fits the data and how the cost decreased over time.

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 6))

# Plot 1: Regression Line
ax1.scatter(X, y, color='blue', label='Data')
ax1.plot(X, predict(X, w, b), color='red', linewidth=2, label='Prediction')
ax1.set_title('Linear Regression Fit')
ax1.set_xlabel('x')
ax1.set_ylabel('y')
ax1.legend()

# Plot 2: Cost History
ax2.plot(cost_history, color='green')
ax2.set_title('Training Cost over Epochs')
ax2.set_xlabel('Epoch')
ax2.set_ylabel('MSE Cost')

plt.show()
```
