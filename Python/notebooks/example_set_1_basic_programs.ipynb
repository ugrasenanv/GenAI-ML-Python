# Python Programming - Example Set 1: Basic Programs

This notebook contains interactive examples of basic Python programming concepts.
Each example demonstrates fundamental programming principles with clear explanations.

## Example 1: Rectangle Area Calculator

```python
# Program to calculate the area of a rectangle
length = float(input("Enter length of the rectangle: "))
width = float(input("Enter width of the rectangle: "))

area = length * width
print("Area of the rectangle is:", area)
```

## Example 2: Simple Interest Calculator

```python
# Program to calculate simple interest
principal = float(input("Enter principal amount: "))
rate = float(input("Enter rate of interest: "))
time = float(input("Enter time (in years): "))

simple_interest = (principal * rate * time) / 100
print("Simple Interest is:", simple_interest)
```

## Example 3: Temperature Conversion

```python
# Program to convert Celsius to Fahrenheit
celsius = float(input("Enter temperature in Celsius: "))

fahrenheit = (celsius * 9/5) + 32
print(f"{celsius}°C is equal to {fahrenheit}°F")
```

## Example 4: Average Calculator

```python
# Program to find the average of three numbers
num1 = float(input("Enter first number: "))
num2 = float(input("Enter second number: "))
num3 = float(input("Enter third number: "))

average = (num1 + num2 + num3) / 3
print("The average is:", average)
```

## Example 5: Square and Cube Calculator

```python
# Program to find the square and cube of a number
num = float(input("Enter a number: "))

square = num ** 2
cube = num ** 3

print("Square:", square)
print("Cube:", cube)
```

## Example 6: Number Swapping

```python
# Program to swap two numbers without using a temporary variable
a = float(input("Enter first number (a): "))
b = float(input("Enter second number (b): "))

# Pythonic way to swap variables simultaneously
a, b = b, a

print("After swapping:")
print("a =", a)
print("b =", b)
```

## Example 7: Student Report Generator

```python
# --- Student Report Program ---

# Step 1: Take student details using input()
student_name = input("Enter student name: ")
roll_number = input("Enter roll number: ")

# Step 2: Store marks in variables
phy_marks = float(input("Enter Physics marks (out of 100): "))
chem_marks = float(input("Enter Chemistry marks (out of 100): "))
math_marks = float(input("Enter Mathematics marks (out of 100): "))

# Step 3: Calculate total and percentage
total_marks = math_marks + chem_marks + phy_marks
percentage = (total_marks / 300) * 100

# Step 4: Display the report card output
print("\n" + "="*30)
print("       STUDENT REPORT CARD      ")
print("="*30)
print(f"Name:        {student_name}")
print(f"Roll No:     {roll_number}")
print("-"*30)
print(f"Total Marks: {total_marks} / 300")
print(f"Percentage:  {percentage:.2f}%")
print("="*30)
```
